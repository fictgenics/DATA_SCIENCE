```python
#import the 2 libraries  - requests and BeautifulSoup
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.amazon.in/Samsung-Galaxy-Bluetooth-Compatible-Android/product-reviews/B0B99SSBWR/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
page = requests.get(url)
page
#Response 200 is fine , Response 503 is an error
#If error persists , try re running the cell for 5 to 10 times or check for proper url
```

```python
page.content # this cell gives us the source code of the amazon review page
#My source code is not is a readable format currently
```

```python
soup = bs(page.content,'html.parser')
```

```python
soup # now we get out html source code in better readable format
```

```python
#ALWAYS IGNORE TOP POSITIVE and TOP CRITICAL reviews
#I just want to extract the names from from reference page
names = soup.find_all('span',class_ = 'a-profile-name')[2:]
names
```

```python
len(names)
```

```python
import pandas as pd
df = pd.DataFrame(names,columns = {'Customer Name'})
df
```

```python
#We are done with names , next is review title
r_title = soup.find_all('a',class_ = 'review-title')
r_title
```

```python
len(r_title)
```

```python
#Data Filtering
review_title = [] #empty title
#for i in range(0,10):
for i in range(0,len(r_title)):
  review_title.append(r_title[i].get_text()[1:-1])#we exclude index 0 and index -1
review_title
```

```python
#Next is review dates
dates = soup.find_all('span',class_ = 'review-date')[2:]
dates
```

```python
#Data filtration
review_date = [] #empty list
for i in range(0,len(dates)):
  review_date.append(dates[i].get_text().replace('Reviewed in India on ',''))
review_date
```

```python
#Next is review rating in stars
rating = soup.find_all('i',class_ = 'review-rating')[2:]
rating
```

```python
len(rating)
```

```python
#data filteration
review_rating =[] #empty list
for i in range(0,len(rating)):
  review_rating.append(rating[i].get_text())
review_rating
```

```python
#Next is review content/review body
content = soup.find_all('span',class_ = 'review-text')
content
```

```python
len(content)
```

```python
#data filtering
review_content =[] # empty list
for i in range(0,len(content)):
  review_content.append(content[i].get_text().replace('\n',''))
review_content
```

```python
df['Review title'] = review_title
df['Review date'] = review_date
df['Review content'] = review_content
df['Review rating'] = review_rating
df
```
